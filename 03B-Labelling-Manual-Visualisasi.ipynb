```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
df = pd.read_csv("data/Data-Test-Manual-Labelling.csv", encoding='latin1')
df
```

```python
print("Akurasi Vader : "+str(len(df[df["vader_labelling_correct_bin"]==1])/len(df)*100)+" %")

conf_mat = confusion_matrix(df["sentimen_manual"], df["auto_sentimen_vader"])
class_label = ["Negatif","Positif"]
test =pd.DataFrame(conf_mat, index = class_label, columns = class_label)
sns.heatmap(test, annot = True, fmt = "d")
plt.title("Confusion Matrix dari Labeling Vader dengan Manual")
plt.xlabel("Data Prediksi")
plt.ylabel("Data Asli")
plt.show()
```
