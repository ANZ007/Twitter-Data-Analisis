```python
import pandas as pd
from nltk.tokenize import word_tokenize
```

```python
positive_lexicon = pd.read_csv('data/lexicon/InSet/positive.tsv',sep='\t')
negative_lexicon = pd.read_csv('data/lexicon/InSet/negative.tsv',sep='\t')
```

```python
lexicon = positive_lexicon.append(negative_lexicon,ignore_index=True)
```

```python
content = (open("data/lexicon/swear-words.txt", "r")).read()
swear_words = content.split("\n")
```

```python
weight_swear = [-5 for i in range(len(swear_words))]
swear_lexicon = pd.DataFrame(list(zip(swear_words,weight_swear)),columns =['word','weight'])
```

```python
add_lexi = pd.read_csv('data/lexicon/agusmakmun/sentiment.csv')
lexi_word = lexicon['word'].to_list()
```

```python
add_word = []
add_weight = []
for i in range(0,len(add_lexi)):
    if (add_lexi['word'][i] not in lexi_word):
        add_word.append(add_lexi['word'][i])
        add_weight.append(add_lexi['weight'][i])

addition_lexicon = pd.DataFrame(list(zip(add_word,add_weight)),columns =['word','weight'])
```

```python
lexicon = lexicon.append(addition_lexicon,ignore_index = True)
lexicon = lexicon.append(swear_lexicon,ignore_index = True)
```

```python
positive_lexicon_2 = pd.read_csv('data/lexicon/ahalimidev/lexicon_positive.csv')
negative_lexicon_2 = pd.read_csv('data/lexicon/ahalimidev/lexicon_negative.csv')
lexicon_2 = positive_lexicon_2.append(negative_lexicon_2,ignore_index=True)
lexi_word_2 = lexicon['word'].to_list()
```

```python
add_word_2 = []
add_weight_2 = []
for i in range(0,len(add_lexi)):
    if (lexicon_2['word'][i] not in lexi_word_2):
        add_word.append(lexicon_2['word'][i])
        add_weight.append(lexicon_2['weight'][i])

addition_lexicon_2 = pd.DataFrame(list(zip(add_word,add_weight)),columns =['word','weight'])
```

```python
lexicon = lexicon.append(addition_lexicon_2,ignore_index = True)
```

```python
def number_of_words(x):
    words = word_tokenize(x['word'])
    number = len(words)
    return number

lexicon['number_of_words'] = lexicon.apply(lambda x: number_of_words(x),axis=1)
lexicon  = lexicon.reset_index(drop=True)
lexicon[lexicon['number_of_words'] == 3]
```
